```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)
data=pd.read_csv('dataset/weight-height.csv')
print(data.shape)
data.head()
```

```python
X = data['Height'].values
Y = data['Weight'].values
```

```python
mean_X = np.mean(X)
mean_Y = np.mean(Y)
l = len(X)
sum1=0
sum2=0
for i in range (l):
    sum1+=(X[i]-mean_X)*(Y[i]-mean_Y)
    sum2+=(X[i]-mean_X)**2
a=sum1/sum2
b=mean_Y-a*mean_X
print(a,b)
```

```python
max_x= np.max(X)+10
min_x= np.max(X)-10
x = np.linspace( min_x ,max_x , 500)
y = a*x+b
plt.plot(x,y,color='green',label='Regression Line')
plt.scatter(X,Y,c='blue',label='Scatter Plot')
plt.xlabel=('heigths')
plt.ylabel=('weights')
plt.legend()
plt.show()
```

```python
summ1=0
summ2=0
for i in range (len(X)):
    summ1+=((a*X[i]+b)-mean_Y)**2
    summ2+=(Y[i]-mean_Y)**2
print(summ1/summ2)    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X=X.reshape((l,1))
reg=LinearRegression()
reg=reg.fit(X,Y)
Y_pred=reg.predict(X)
print(reg.score(X,Y))

```
