```python
import pandas as pd 
dataset=pd.read_csv("https://cocl.us/datascience_survey_data")
dataset.head()
```

```python
sortedData=dataset.sort_values(by='Very interested', ascending=False)
sortedData.head()
```

```python
#Convert the numbers into percentages

sortedData['Very interested']= (100*sortedData['Very interested']/2233.0).round(2)
sortedData['Somewhat interested']= (100*sortedData['Somewhat interested'] /2233.0).round(2)
sortedData['Not interested']= (100*sortedData['Not interested'] /2233.0).round(2)
sortedData.head()
```

```python
import matplotlib.pyplot as plt

pl=sortedData.plot(kind='bar', figsize=(20, 8) , width=0.8, color=['#5cb85c','#5bc0de','#d9534f'] ,fontsize=14)
pl.set_xticklabels(sortedData["Unnamed: 0"] ,fontsize=14)
#remove the left, top, and right borders
pl.spines['right'].set_visible(False)
pl.spines['left'].set_visible(False)
pl.spines['top'].set_visible(False)
#remove the y axe
plt.yticks([])
#display the percentages above the bars
for i in pl.patches:
    pl.text(i.get_x(), i.get_height()+3, str(i.get_height())+'%', fontsize=14)
#add title    
plt.title('percentage of the respondents interest in Data Science areas', fontsize=16)
plt.show()
```
