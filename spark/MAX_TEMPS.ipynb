```python
import findspark
findspark.init('C:\\Spark')
findspark.find()
```

```python
from pyspark import SparkConf , SparkContext
```

```python
conf = SparkConf().setMaster ("local").setAppName("Temperature")
sc = SparkContext (conf = conf)
```

```python
def parseLine(line):
    fields = line.split(',')
    DATE=fields[1]
    type_entree =fields[2]
    temperature =float(fields[3])*0.1*(9.0/5.0) +32.0
    return(DATE ,  type_entree ,temperature)
```

```python
lines = sc.textFile("C:/Users/meher/Downloads/datasets/temperatures.csv")
parsedLines=lines.map(parseLine)
```

```python
maxTemps = parsedLines.filter(lambda x: "TMAX" in x[1])
```

```python
dateTemps = maxTemps.map(lambda x:(x[0], x[2]))
```

```python
maxTemps = dateTemps.reduceByKey(lambda x,y:(max(x,y)))   
```

```python
results2 = maxTemps.collect()
```

```python
for result in results2:
    print(result[0] + "\t{:.3}".format(result[1]))
```

```python
max_temp = max(results2, key = lambda x:x[1])
max_temp[1]
```
